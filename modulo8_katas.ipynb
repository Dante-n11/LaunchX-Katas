# Ejercicio 1: Creación de diccionarios de Python

```python
planet = {
    'name': 'Mars',
    'moons': 2
}
print(f"Nombre: {planet.get('name')}")
print(f"Numero de lunas: {planet['moons']}\n")

planet['circunferencia (km)'] = {
    'polar': 6752,
    'equatorial': 6792
}

print(f"Nombre: {planet['name']}")
print(f"Circunferencia polar: {planet['circunferencia (km)']['polar']}")
```

# Ejercicio 2: Programación dinámica con diccionarios

```python
planet_moons = {
    'mercury': 0,
    'venus': 0,
    'earth': 1,
    'mars': 2,
    'jupiter': 79,
    'saturn': 82,
    'uranus': 27,
    'neptune': 14,
    'pluto': 5,
    'haumea': 2,
    'makemake': 1,
    'eris': 1
}

noPlanetas = len(planet_moons.keys())
noTotalLunas = 0
for lunasXPlaneta in planet_moons.values() :
    noTotalLunas += lunasXPlaneta

promedioLunas = noTotalLunas / noPlanetas

print(f'El numero total de lunas es: {noTotalLunas}')
print(f'El numero promedio de lunas por planeta es: {promedioLunas}')
```
