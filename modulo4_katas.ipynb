# Ejercicio 1: Transformar cadenas

```python
text = """Interesting facts about the Moon. The Moon is Earth's only satellite. There are several interesting facts about the Moon and how it affects life here on Earth. On average, the Moon moves 4cm away from the Earth every year. This yearly drift is not significant enough to cause immediate effects on Earth. The highest daylight temperature of the Moon is 127 C."""

#Dividiento el texto en cada oración
textoDividido = text.split('. ')

palabrasClave = ['average', 'temperature', 'distance']

for oracion in textoDividido :
    for palabra in palabrasClave :
        if palabra in oracion :
            print (oracion)

#Cambiando 'C' por 'Celsius'
print()
for oracion in textoDividido :
    for palabra in palabrasClave :
        if palabra in oracion :
            print (oracion.replace('C', 'Celsius'))
```

# Ejercicio 2: Formateando cadenas

```python
#Datos originales
name = "Moon"
gravity = 0.00162 #(kms)
planet = "Earth"

titulo = f"gravity facts about {name}"
contenido = (f"""
{'-'*60}
Planet Name: {planet}
Gravity on Ganymede: {gravity*1000} m/s2""")

plantilla = f"{titulo.title()}{contenido}\n\n"

print(plantilla)


#Nuevos datos
name = "Marte"
gravity = 0.00143
planet = "Ganímedes"

print(plantilla)


#Nueva plantilla aplicando format()
plantillaNueva = ("""Gravity Facts About {name}
------------------------------------------------------------
Planet Name: {planet}
Gravity on Ganymede: {gravity} m/s2""")

print(plantillaNueva.format(name=name, planet=planet, gravity=gravity*1000))
```
