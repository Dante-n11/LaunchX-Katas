# Ejercicio 1: Trabajar con argumentos en funciones

```python
def informe(tanque1, tanque2, tanque3):
    prom = (tanque1+tanque2+tanque3) / 3
    return (f"El promedio es: {prom}")

print(informe(50, 70, 60))


def promedio(*tanques) :
    return (sum(tanques) / len(tanques))


def informeMejorado(tanque1, tanque2, tanque3):
    prom = promedio(tanque1, tanque2, tanque3)
    return (f"El promedio es: {prom}")

print(informeMejorado(50, 70, 60))
```

# Ejercicio 2: Trabajo con argumentos de palabra clave

```python
def informePreciso (horaPre, tiempoV, destino, tanqueExt, tanqueInt):
    return (f"""
    El destino es: {destino}
    El tiempo de vuelos es: {horaPre+tiempoV} min
    El total de combustible es: {tanqueExt+tanqueInt} galones""")

print(informePreciso(12, 45, 'Luna', 150000, 180000))


def informePreciso2 (destino, *minutos, **tanques):
    return (f"""
    El destino es: {destino}
    El tiempo de vuelos es: {sum(minutos)} min
    El total de combustible es: {sum(tanques.values())} galones""")

print(informePreciso2('Luna', 12, 45, tanqueExt=150000, tanqueInt=180000))


def informePreciso3 (destino, *minutos, **tanques):
    informe = f"""
    El destino es: {destino}
    El tiempo de vuelos es: {minutos[0]+minutos[1]} min
    El total de combustible es: {sum(tanques.values())} galones"""
    for tipoTanque, combustible in tanques.items():
        informe += f"""
        *{tipoTanque}: {combustible}"""

    return informe

print(informePreciso3('Luna', 12, 45, tanqueExt=150000, tanqueInt=180000))
```
