# Ejercicio1: Crear y usar listas de Python

```python
planets = ['Mercurio', 'Venus', 'Tierra', 'Marte', 'Jupiter', 'Saturno', 'Urano', 'Neptuno']
print ('El numero de planetas es:',len(planets))

planets.append('Plutón')
print ('El numero de planetas es:', len(planets))
print ('El último planeta es:', planets[-1])
```

# Ejercicio 2: Trabajando con datos de una lista

```python
planetas = ['Mercurio', 'Venus', 'Tierra', 'Marte', 'Jupiter', 'Saturno', 'Urano', 'Neptuno']
planetaUsuario = input('Ingresa el nombre de un planeta colocando el nombre con mayuscula')

posicion = planetas.index(planetaUsuario)
print('Posicion del planeta ingresado: '+str(posicion))

planetasMasCercanos = planetas[0:posicion]
print('Planetas más cercanos al sol: '+str(planetasMasCercanos))

planetasMasLejanos = planetas[posicion+1:8]
print('Planetas más lejanos al sol: '+str(planetasMasLejanos))
```
