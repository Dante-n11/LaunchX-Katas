# Ejercicio 1: Escribir declaraciones if, else y elif

```python
velocidad = 49

if velocidad > 25 :
    print('¡Advertencia! el asteroide se acerca demasiado rapido ('+str(velocidad)+' km/s)')
else :
    print('La velocidad del asteroide no causa ningún peligro')
```

```python
velocidadEntrada = 19

if velocidadEntrada > 20 :
    print('Se puede visualizar un rayo de luz')
elif velocidadEntrada == 20 :
    print('Se puede visualizar un rayo de luz')
else : 
    print('No se puede visualizar ningun rayo de luz')
```

# Ejercicio 2: Operadores and y or

```python
tamanio = 20
velocidad2 = 15

if tamanio > 25 and velocidad2 > 25 :
    print('El tamaño del asteroide puede causar muchos daños')
    print('¡Advertencia! el asteroide se acerca demasiado rapido')
elif velocidad2 >= 20 :
    print('Se puede visualizar un rayo de luz')
elif tamanio < 25 :
    print('El tamaño del asteroide no causa problemas')
else : 
    print ('Todo esta en orden')
```
